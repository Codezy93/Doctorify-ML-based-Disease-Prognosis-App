```python
import spacy
import scispacy
from scispacy.umls_linking import UmlsEntityLinker
```

```python
def extract_medical_entities(text):
    nlp = spacy.load("en_core_sci_lg")  # Load a scispaCy model
    linker = UmlsEntityLinker(resolve_abbreviations=True)  # UMLS Linking
    nlp.add_pipe('scispacy_linker', last=True, config={"linker": linker})

    doc = nlp(text)
    entities = []
    for ent in doc.ents:
        for umls_ent in ent._.umls_ents:
            cui, _ = umls_ent
            entity = linker.umls.cui_to_entity[cui]
            if 'T047' in entity.types:  # T047 is the UMLS type for diseases/disorders
                entities.append((ent.text, entity.canonical_name))
    return entities
```

```python
input_text = "The patient has pain in the head and is suffering from short breath."

detected_entities = extract_medical_entities(input_text)
print("Detected Medical Entities:", detected_entities)
```
